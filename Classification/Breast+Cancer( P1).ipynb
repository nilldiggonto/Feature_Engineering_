<h3 class="text-center"> Importing Library </h3>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(palette='rainbow')
```

```python
import warnings
warnings.filterwarnings(action='ignore')
```

<h4 class="text-center"> LOAD DATA FROM SCIPY </h4>
<br>

```python
from sklearn.datasets import load_breast_cancer
```

```python
raw_data = load_breast_cancer()
```

```python
raw_data.keys()
```

```python
raw_data['data'].shape
```

```python
raw_data['feature_names']
```

```python
raw_data['target_names']
```

##### Creating a dataFrame

```python
data = pd.DataFrame(np.c_[raw_data['data'],raw_data['target']], columns= np.append(raw_data['feature_names'],['target']))
```

```python
data.head()
```

```python
data.tail()
```

```python
data.shape
```

<h4 class="text-center"> VISUALIZE DATA </h4>
<br>

```python
sns.pairplot(data, vars =['mean radius', 'mean texture', 'mean perimeter', 'mean area'] ,hue='target',palette='Set1');
# 0 ='malignant', 1= 'benign'
```

```python
#### Checking the distribution of 'malignant', 'benign' cancer
```

```python
data['target'].value_counts() #(0 means malignant and 1 means benign)
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data['target'],palette='Set1');
plt.xlabel('Malignant & Benign');
```

```python
#checking the corelation
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(data.corr(),annot=True,cbar=False,cmap='rainbow');
```

<h4 class="text-center"> ML MODEL </h4>
<br>

<span class='badge'> TRAIN TEST SPLIT </span>

```python
from sklearn.model_selection import train_test_split
```

```python
featured_x = data.drop('target',axis=1)
target_y = data['target']
```

```python
x_train,x_test,y_train,y_test = train_test_split(featured_x,target_y,test_size=0.2,random_state=123)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

<span class="badge"> SVM </span>

```python
from sklearn.svm import SVC
```

```python
clf = SVC()
```

```python
clf.fit(x_train,y_train)
```

```python
clf.score(x_train,y_train)
```

<span class="badge"> Prediction </span>

```python
y_preds = clf.predict(x_test)
```

```python
y_preds
```

<span class='badge'> Confusion Matrix </span>

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_preds)
```

```python
sns.heatmap(confusion_matrix(y_test,y_preds),annot=True,cbar=False,cmap='gist_gray');
plt.ylabel('Predicted value')
plt.xlabel('Actual value')
plt.show()
```

<span class='badge'> Classification Report </span>

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_preds))
```

<h4 class="text-center">EVALUATE ML MODEL </h4>
<br>

<span class='badge'> min-max </span>

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler =  MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
scaled_feature = scaler.transform(x_train)
```

```python
#fit with new scaled data
clf.fit(scaled_feature,y_train)
```

<span class="badge"> Predict </span>

```python
scaled_test_feature = scaler.fit_transform(x_test)
```

```python
y2_preds = clf.predict(scaled_test_feature)
```

<span class='badge'> Confusion Matrix </span>

```python
confusion_matrix(y_test,y2_preds)
```

```python
sns.heatmap(confusion_matrix(y_test,y2_preds),annot=True,cbar=False,cmap='gist_gray');
plt.ylabel('Predicted value')
plt.xlabel('Actual value')
plt.show()
```

```python
print(classification_report(y_test,y2_preds))
```
